```python
#Lists  that require tow indices to identify an element are called TWO-DIMENSIONAL LISTS
```

```python
#Creating a two-dimensional list
a = [[77, 68, 86, 73], [96, 87, 89, 81], [70, 90, 86, 81]]
```

```python
for i in a:
    print(i, end="")
```

```python
#How the Nested Loops Execute
for i, row in enumerate(a):
    for j, item in enumerate(row):
        print(f'a[{i}][{j}]={item}', end=" ")
    print()
```

```python
t = [[10, 7, 3], [20, 4, 17]]
total = 0
items = 0
for rows in t:
    for i in rows:
        total += i
        items += 1
print(total/items)
```

```python
total = 0
items = 0
for row in t:
    total += sum(row)
    items += len(row)

print(total/items)
```
